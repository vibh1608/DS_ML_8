```python
import nltk   #library for natural language processing (NLP)
import matplotlib.pyplot as plt  #library for data visualization
```

```python
#library to extract text from webpage.
import urllib.request

#sending request to get the web page content
response = urllib.request.urlopen('http://php.net/')

#getting response from the request
html = response.read()

#printing the html content
print (html)
```

### As it is clear from printed output, the result contains a lot of HTML tags that need to be cleaned. We can use BeautifulSoup to clean the grabbed text.

```python
#library for parsing HTML and XML documents(text)
from bs4 import BeautifulSoup

#gremoving all tag of html content and storing the result in soup variable
soup = BeautifulSoup(html,"html5lib")

#getting all cleaned text into text variable
text = soup.get_text(strip=True)

#priting the clean text from the crawled web page.
print (text)
```

```python
#converting that text into tokens by splitting the text
tokens = [t for t in text.split()] 

#printing the tokens
print (tokens)
```

```python
#finding frequency of each token using FreqDist function
freq = nltk.FreqDist(tokens) 

#printing each token with it's frequency
for key,val in freq.items(): 
    print (str(key) + ':' + str(val))
```

```python
#plotting the graph between each token and it's frequency
freq.plot(20, cumulative=False)

#showing the plot
plt.show()
```
